```
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style='darkgrid'
```

# BEFORE EXTRA TWEAKING

```
df = pd.read_csv('/content/drive/MyDrive/Group Project - Kickstarter/models.csv')
```

```
df.head()
```

```
plt.subplots(figsize=(14,8))
colors = ["#97AABD","#C96567"]
sns.set_palette(sns.color_palette(colors))
sns.scatterplot(data=df, x='model', y='accuracy', hue='Model Version', s=100)
sns.set_theme(context='notebook', style='darkgrid')
plt.title('Performance of Machine Learning Models', fontsize=20, pad=20)
plt.xlabel('', labelpad=10, fontsize=15)
plt.ylabel('Model Accuracy (%)', labelpad=10, fontsize=15)
plt.legend(loc='lower right', prop={'size':20})
```
